```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combine_df = pd.merge(city_df, ride_df, on='city', how='outer')

# Display the data table for preview
combine_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
type_urban =  combine_df.loc[combine_df['type']== "Urban"]
type_rural  = combine_df.loc[combine_df["type"] == "Rural"]
type_sub = combine_df.loc[combine_df['type'] == "Suburban"]

ride_count_urban = type_urban.groupby(['city'])['ride_id'].count()
ride_count_rural = type_rural.groupby(['city'])['ride_id'].count()
ride_count_sub = type_sub.groupby(['city'])['ride_id'].count()

driver_count_urban = type_urban.groupby(['city'])['driver_count'].count()
driver_count_rural = type_rural.groupby(['city'])['driver_count'].count()
driver_count_sub = type_sub.groupby(['city'])['driver_count'].count()

average_urban = type_urban.groupby(['city'])['fare'].mean()
average_rural = type_rural.groupby(['city'])['fare'].mean()
average_sub = type_sub.groupby(['city'])['fare'].mean()

# Build the scatter plots for each city types
plt.scatter(ride_count_urban, average_urban , s=10*driver_count_urban, marker='o', facecolors='coral',edgecolors='black', label='Urban')
plt.scatter(ride_count_rural, average_rural, s=10*driver_count_rural, marker='o', facecolors='gold',edgecolors='black', label='Rural')
plt.scatter(ride_count_sub, average_sub, s=10*driver_count_sub , marker='o', facecolors='skyblue',edgecolors='black', label='Suburban')
plt.grid()
plt.legend(loc='best', title = "City Types")
plt.xlabel('Total Number of Ridesd (per city)')
plt.ylabel('Average Fare(s)')
plt.title('Pyber Ride Showing Data(2016)')
plt.savefig('../Images/Pyber_Ride')
plt.show()

```

## Total Fares by City Type

```python
# Calculate Type Percents

grp_driver =  combine_df.groupby(["type"])
per_total_far = grp_driver['fare'].sum()
label = per_total_far.keys()
colors =["gold", "lightskyblue" , "lightcoral"]
per_total_far.plot(kind='pie', y=label, autopct='%1.1f%%', explode = (0,0,0.1), startangle=180,
                  title='% of Total Fares by City Type', colors=colors, shadow=True)

plt.savefig('../Images/Total_Fares.png')
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_ride_count = grp_driver['ride_id'].count()
label2 =  total_ride_count.keys()
colors =["gold", "lightskyblue" , "lightcoral"]
total_ride_count.plot(kind='pie', y=label2, autopct='%1.1f%%', explode = (0.1,0,0), title='% of Total Rides by City Type', colors=colors,
                     shadow=True)

plt.savefig('../Images/Total_Rides')
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
city_grp = city_df.groupby(["type"])
driver_count = city_grp['driver_count'].sum()
colors =["gold", "lightskyblue" , "lightcoral"]

label3 = driver_count.keys()
driver_count.plot(kind='pie', y=label3, autopct='%1.1f%%', explode = (0,0.1,0), title='% of Total Drivers by City Type', startangle=180,
                 colors=colors, shadow=True)

plt.savefig('../Images/Total_Drivers')
plt.show()
```
